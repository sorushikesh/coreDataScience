### Variance
- Measures the **average squared deviation** from the mean.
- High variance = data spread out; Low variance = data close to the mean.
- Formula:
\[
\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2
\]

### Standard Deviation
- Square root of the variance.
- Represents **average distance** from the mean.
- Same units as the data.
- Formula:
\[
\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2}
\]


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Style
sns.set(style="whitegrid")
```

```python
# Generate two datasets: one with low spread, one with high spread
np.random.seed(42)
data_low_var = np.random.normal(50, 5, 1000)
data_high_var = np.random.normal(50, 25, 1000)

df = pd.DataFrame({
    "Low Variance": data_low_var,
    "High Variance": data_high_var
})
df.head()
```

```python
# Low variance data
var_low = np.var(data_low_var)
std_low = np.std(data_low_var)

# High variance data
var_high = np.var(data_high_var)
std_high = np.std(data_high_var)

print(f"Low Variance Data - Variance: {var_low:.2f}, Std Dev: {std_low:.2f}")
print(f"High Variance Data - Variance: {var_high:.2f}, Std Dev: {std_high:.2f}")
```

```python
plt.figure(figsize=(12, 6))

sns.kdeplot(data_low_var, label=f"Low Variance (={std_low:.2f})", fill=True)
sns.kdeplot(data_high_var, label=f"High Variance (={std_high:.2f})", fill=True)

plt.title("Distributions with Low and High Variance")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
