# Box and Whisker Plot (Box Plot)

A **box and whisker plot** is a standardized way of displaying the distribution of data based on five summary statistics:

---

## Five Number Summary
1. **Minimum** – the smallest value (excluding outliers)
2. **Q1 (First Quartile)** – 25th percentile
3. **Median (Q2)** – 50th percentile
4. **Q3 (Third Quartile)** – 75th percentile
5. **Maximum** – the largest value (excluding outliers)

Outliers are plotted as individual points.

---

## When to Use
- To understand **data distribution**
- To detect **outliers**
- To compare distributions across multiple groups

---


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Generate sample data
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=100)

# Box plot
plt.figure(figsize=(8, 5))
plt.boxplot(data, vert=False)
plt.title('Box and Whisker Plot')
plt.xlabel('Value')
plt.show()
```

```python
# Multiple group data
group_A = np.random.normal(60, 8, 100)
group_B = np.random.normal(50, 10, 100)
group_C = np.random.normal(40, 6, 100)
```

```python

# Boxplot
plt.figure(figsize=(10, 6))
plt.boxplot([group_A, group_B, group_C], labels=['Group A', 'Group B', 'Group C'])
plt.title('Box Plot for Multiple Groups')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
# Create DataFrame
df = pd.DataFrame({
    'Group': ['A'] * 100 + ['B'] * 100 + ['C'] * 100,
    'Value': np.concatenate([group_A, group_B, group_C])
})

# Boxplot using seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Value', data=df)
plt.title('Box Plot with Seaborn')
plt.show()
```

## Box Plot in Machine Learning

- Detect **outliers** in feature values.
- Visualize **feature distributions**.
- Compare **target variable** across categories.
-
- Outliers: Typically shown as dots outside whiskers.
- Whiskers usually extend to 1.5×IQR from Q1 and Q3.
- Use box plots during **EDA (Exploratory Data Analysis)** phase.
