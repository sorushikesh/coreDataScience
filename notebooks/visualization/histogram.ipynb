# Histogram

A **histogram** is a graphical representation that organizes a group of data points into user-specified ranges (called bins). It is used to visualize the distribution of numerical data.

---

## Key Components
- **Bins**: Ranges that group the data
- **Height of bars**: Frequency (count) of data in each bin
- **X-axis**: Range of values
- **Y-axis**: Frequency count

---

## When to Use
- To understand the **shape** of the data distribution
- To check for **skewness**, **modality**, and **outliers**
- During **Exploratory Data Analysis (EDA)**

---


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
# Generate sample data
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)
```

```python
# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, edgecolor='black', color='skyblue')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# More granular histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=[30, 40, 50, 60, 70], edgecolor='black', color='orange')
plt.title('Histogram with Custom Bins')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Seaborn version
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=20, kde=True, color='green')
plt.title('Histogram with KDE (Seaborn)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

## Histogram vs Bar Plot

| Feature                       | Histogram                                   | Bar Plot                           |
|-------------------------------|---------------------------------------------|------------------------------------|
| Data Type                     | Numerical (continuous) on X-axis            | Categorical on x-axis              |
| Gaps Between Bars             | No (bars touch)                             | Yes (bars are separated)           |
| X-axis                        | Binned numeric ranges                       | Categories or labels               |
| Purpose                       | Show distribution of data                   | Compare values across categories   |
| Bin Width                     | Can be adjusted                             | Not applicable                     |

---

### When to Use
- **Use Histogram** for analyzing data **distribution**, **skewness**, and **outliers** in continuous variables.
- **Use Bar Plot** for comparing **frequencies** or **aggregated metrics** across discrete categories.

```python
# Read data from CSV file
df = pd.read_csv("../../data/cristiano_goals.csv")

# Filter for Manchester United and UEFA Champions League
df = df[(df["Club"] == "Manchester United") & (df["Competition"] == "UEFA Champions League")]

plt.figure(figsize=(10, 6))
sns.histplot(df["Season"], bins=15, kde=True, color="royalblue")

plt.title("Cristiano Ronaldo  UCL Goals for Manchester United by Season")
plt.xlabel("Season")
plt.ylabel("Number of Goals")
plt.grid(True)
plt.show()
```

```python
# Read data
df = pd.read_csv("../../data/cristiano_goals.csv")

# Filter for UEFA Champions League
df = df[df["Competition"] == "UEFA Champions League"]

# Plot histogram with different colors for each club
plt.figure(figsize=(12, 6))
custom_colors = {"Real Madrid": "navy", "Juventus FC": "black", "Manchester United": "red"}
sns.histplot(data=df, x="Season", hue="Club", multiple="stack", palette=custom_colors)

plt.title("Cristiano Ronaldo  UCL Goals by Season (Grouped by Club)")
plt.xlabel("Season")
plt.ylabel("Number of Goals")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
