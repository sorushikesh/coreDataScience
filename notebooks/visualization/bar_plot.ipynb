# Bar Plot in Statistics and Machine Learning

A **bar plot** is a chart that uses rectangular bars to represent categorical data.
It's useful for comparing the frequency or value of different categories.

---

## Characteristics
- Categories on **x-axis**
- Frequencies or values on **y-axis**
- Bars can be **vertical** or **horizontal**

---

## When to Use
- To compare **counts** or **aggregated values**
- To visualize **categorical variables**
- Great for **EDA (Exploratory Data Analysis)**

---


```python
import matplotlib.pyplot as plt

categories = ["Real Madrid FC", "Atletico Madrid FC", "FC Barcelona", "Athletic Bilbao",
              "Real Sociedad", "Deportivo La Coruna", "Sevilla", "Valencia"]

values = [36, 11, 28, 8, 2, 1, 1, 6]

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(categories, values, align='center')
plt.title("La Liga Title Winners")
plt.xlabel("La Liga Teams")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ["Real Madrid FC", "Atletico Madrid FC", "FC Barcelona", "Athletic Bilbao",
              "Real Sociedad", "Deportivo La Coruna", "Sevilla", "Valencia"]
values = [36, 11, 28, 8, 2, 1, 1, 6]

# Sort by number of titles (descending)
sorted_data = sorted(zip(values, categories), reverse=True)
sorted_values, sorted_categories = zip(*sorted_data)

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(sorted_categories, sorted_values)
plt.title("La Liga Title Winners (Sorted)")
plt.xlabel("La Liga Teams")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45, ha='right')

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.3, str(height),
             ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
