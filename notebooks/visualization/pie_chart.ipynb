# Pie Chart

A **pie chart** is a circular statistical graphic divided into slices to illustrate **proportional data**.

Each slice represents a **category's contribution to the whole**.

---

## When to Use

- To show **percentage or proportional data**
- Tto **compare parts of a whole**
- When categories are **mutually exclusive**

---

## Characteristics

- Each sliceâ€™s size is proportional to the **category's frequency or sum**
- Best used when there are **fewer categories** (ideally < 6)
- Often annotated with **percentage values**

---

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV
df = pd.read_csv("../../data/cristiano_goals.csv")

# Filter for UEFA Champions League
ucl_df = df[df["Competition"] == "UEFA Champions League"]

# Group by club and count goals
goal_counts = ucl_df["Club"].value_counts()

# Pie chart
plt.figure(figsize=(10, 6))
plt.pie(
    goal_counts, 
    labels=goal_counts.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    wedgeprops={'edgecolor': 'white'}
)

plt.title("Cristiano Ronaldo â€“ UCL Goals by Club")
plt.axis('equal')
plt.show()
```
